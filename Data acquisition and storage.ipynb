# Data acquisition

Using *yfinance* library to download data of APPLE stock price, Gold, Crude oil, Dow Jones Industrial Average and NASDAQ Index from Yahoo! finance starting from 2017-04-01 to 2020-05-31.

```python
import yfinance as yf

apple = yf.download("AAPL", start="2017-04-01", end="2020-05-31", group_by="ticker")
gold = yf.download("GC=F", start="2017-04-01", end="2020-05-31", group_by="ticker")
oil = yf.download("CL=F", start="2017-04-01", end="2020-05-31", group_by="ticker")
dj = yf.download("^DJI", start="2017-04-01", end="2020-05-31", group_by="ticker")
nd = yf.download("^NDX", start="2017-04-01", end="2020-05-31", group_by="ticker")

apple
```

Downloading ratings of APPLE company using *pyfmpcloud* library.

```python
from pyfmpcloud import company_valuation as cv

rating = cv.rating('AAPL', history = 'daily')
rating
```

Retrieving tweets containing #APPL from Twitter using *twint* library.

```python
import twint
import nest_asyncio
nest_asyncio.apply()

c = twint.Config()

c.Search = "#AAPL"
c.Since = '2017-04-01'
c.Until = '2020-05-31'
c.Pandas = True
c.Lang = 'en'
c.Store_csv=True
c.Output="./dataset/apple_twitter.csv"

# Run
twint.run.Search(c)
```

```python
import pandas as pd
import numpy as np
import re
import nltk

raw = pd.read_csv('./dataset/apple_twitter.csv')

df = raw[['date', 'tweet']]

# convert to lower string
df['tweet'] = df['tweet'].str.lower()

# convert into words
for i in range(len(df['tweet'])):
    df['tweet'][i] = nltk.word_tokenize(df['tweet'][i])

# covert to string 
for i in range(len(df['tweet'])):
    df['tweet'][i] = str(df['tweet'][i])

# remove special words
for i in range(len(df['tweet'])):
    df['tweet'][i] = re.sub('[^ a-zA-Z0-9]', '', df['tweet'][i])

# aggregate tweets of each day into bag of words
df1 = df.groupby(['date'], as_index = False).agg({'tweet': ','.join})

# remove special words
for i in range(len(df1['tweet'])):
    df1['tweet'][i] = re.sub('[^ a-zA-Z0-9]', '', df1['tweet'][i])
df1
```

```python
# get sentimental score of each day

df1['Comp']=""
df1['Negative']=""
df1["Positive"]=""
df1['Neutral']=""

# converting tweets into string and remove brackets
for i in range(len(df1['Tweet'])):
    df1['Tweet'][i]=str(df1['Tweet'][i])

from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

# calculating values from tweets of each day
for i in range(len(df1['Tweet'])):
    value = sid.polarity_scores(df1['Tweet'][i])
    df1['Comp'][i] = value['compound']
    df1['Negative'][i] = value['neg']
    df1['Positive'][i] = value['pos']
    df1['Neutral'][i] = value['neu']
df1
```
